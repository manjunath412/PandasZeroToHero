```python
# Merge master and scoring dfs on 'master' index and 'scoring' playerID column
# left_index=True: This parameter tells the function to use the index of the left DataFrame master 
# as the join key. Instead of looking for a common column to join on, it uses the index of the master DataFrame.
# right_on="playerID": This parameter specifies that the playerID column in the right DataFrame scoring
# should be used as the join key. This means that the function will look for matching values between the 
# index of the master DataFrame and the playerID column of the scoring DataFrame.
#print(master.index.dtype)
#print(scoring['playerID'].dtype)
#scoring['playerID'] = scoring['playerID'].astype('int64')
#master.index = master.index.astype(str)
pd.merge(master, scoring, left_index=True, right_on="playerID").head()
```

```python
# set player_id column of scoring df as index, drop the column, and join on index of the 2 tables
# since both tables now have playerid as index, they will join based on the shared index
pd.merge(master, scoring.set_index("playerID", drop=True),
                                   left_index=True, right_index=True).head()
```

```python
# Resetting the index to default and removing the old index
scoring = scoring.reset_index(drop=True)
# Merging 'master' with 'scoring' using the index of 'master' and 'playerID' from 'scoring'
pd.merge(master, scoring, left_index=True,
                  right_on="playerID").head()
```

```python
print(

    pd.merge(master, scoring, left_index=True,
             right_on="playerID").shape,
    
    pd.merge(master, scoring, left_index=True,
                  right_on="playerID", how="right").shape
)
```

```python
# Drop random records
#difference between right join versus inner join, right join will bring all in scoring, and 
# nans for not matching in masters2
master2 = master.drop(master.sample(5).index)

print(

    pd.merge(master2, scoring, left_index=True,
             right_on="playerID").shape,
    
    pd.merge(master2, scoring, left_index=True,
                  right_on="playerID", how="right").shape
)
```

```python
# right join with playerid from scoring and index from master2
# "indicator" tracks the source of each row, showing whether the data comes from 'master2', 'scoring', or both.
merged = pd.merge(master2, scoring, left_index=True,
                  right_on="playerID", how="right", indicator=True)
merged.head()
```

```python
# how many from scoring and how many from right only versus left only
merged["_merge"].value_counts()
```

```python
# shows right join where there is a scoring record and no matching from master
merged[merged["_merge"] == "right_only"].head() #filter where clause on _merge column
```

```python
# Dropping a random sample of 1000 rows from 'scoring', does not modify in-place, therefore assign to another df
scoring2 = scoring.drop(scoring.sample(1000).index)
```

```python
merged = pd.merge(master2, scoring2, left_index=True,
                  right_on="playerID",
                  how="outer", indicator=True)
```

```python
merged[(merged["_merge"] == "left_only") |
       (merged["_merge"] == "right_only")
      ].sample(3)
```
