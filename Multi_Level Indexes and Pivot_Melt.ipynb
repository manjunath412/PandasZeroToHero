```python
import numpy as np
```

```python
# for writing complex reusable logic to get locations for filtering in mult-index, variations
# can be used in regular dfs also
def get_many_locs(df, slices):
    arr = np.empty(0, dtype="int")
    for s in slices:
        locs = df.index.get_locs((s))
        arr = np.concatenate((arr, locs))
    return arr
    
```

```python
locs = get_many_locs(
           mi,
           (
               idx['aaltoan01':'adamscr01', 1997:2000],
               idx['aaltoan01':'adamscr01', 2004:2006]
           ))
locs
```

```python
sliced = mi.iloc[locs, :] # using the locs from above where we are having multiple query conditions
sliced.head(10)
```

# Wide (Pivot) and Long (Melt) Formats Revisited

```python
team_splits.head(5)
```

```python
team_splits = team_splits.drop(columns=["tmID"])
team_splits = team_splits.set_index(["year", "name"])
team_splits.head()
```

```python
months = team_splits.columns.map(lambda x: x[:3]) # lambda is a fancy function, split column name to 1st 3 chars
metrics = team_splits.columns.map(lambda x: x[3:])
print(months)
print(metrics)
```

```python
midx = pd.MultiIndex.from_arrays([months, metrics])
midx
```

```python
team_splits.columns = midx
team_splits.head() #better readability using multi-index levels and columns subbed into Win, Loss, Tie
```

```python
team_splits = team_splits.stack(level=0)
team_splits.head(10)
```

```python
team_splits = team_splits.swaplevel(1, 2)
team_splits.head()
```

```python
team_splits = team_splits.reorder_levels([2,0,1])
team_splits.head()
```

```python
#team_splits.index.levels[2].name = "month"
team_splits.index.set_names('month', level=2, inplace=True)
team_splits.head()
```

```python
team_splits.groupby(level=["name", "year"]).sum().head(10)
```

```python
team_splits.unstack(level=["year", "month"]).head()
```
